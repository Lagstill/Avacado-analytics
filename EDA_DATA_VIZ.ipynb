### It appears that the data of Children in employment (both male and female) have about 98% and Literacy Rate of about 84% of zero-entries . Further, we will be considering the total Refugee population (asylum,country) and drop the redundant features.###

```python
df['Refugee population']=df['Refugee population by country or territory of asylum']+df['Refugee population by country or territory of origin']
df= df.drop(['Labor force participation rate, total (% of total population ages 15-64) (modeled ILO estimate)','Country Name','Children in employment, work only, male (% of male children in employment, ages 7-14)','Children in employment, work only, female (% of female children in employment, ages 7-14)','Labor force participation rate, male (% of male population ages 15-64) (modeled ILO estimate)','Labor force participation rate, female (% of female population ages 15-64) (modeled ILO estimate)','Labor force participation rate for ages 15-24, male (%) (modeled ILO estimate)','Labor force participation rate for ages 15-24, female (%) (modeled ILO estimate)','Labor force participation rate for ages 15-24, total (%) (modeled ILO estimate)','Children in employment, study and work (% of children in employment, ages 7-14)','Prevalence of HIV, female (% ages 15-24)','Prevalence of HIV, male (% ages 15-24)','Children in employment, study and work, female (% of female children in employment, ages 7-14)','Children in employment, study and work, male (% of male children in employment, ages 7-14)','Children in employment, total (% of children ages 7-14)','Children in employment, work only (% of children in employment, ages 7-14)','Refugee population by country or territory of origin','Refugee population by country or territory of asylum','School enrollment, primary and secondary (gross), gender parity index (GPI)'], axis=1)
```

### Further any rows containing zero-entries are removed and that will be considered as the final data ###

```python
nan=df[(df != 0)].dropna()
cleaned_data=nan.sort_values(by='Year')
cleaned_data
```

#### These countries and the year will be our area of interest.

```python
print(cleaned_data['Country Code'].unique())
print(cleaned_data['Year'].unique())
```

```python
cleaned_data.columns
```

### Now, we will seggregate the data  for analysis by country and year.  *(Note the structure of the data ,for each country there is 3-5 year (e.g: 2000,2004,2015) data entry ,vice-versa.Hence, we consider the average of it , rejecting nothing.)*  ###




```python
country_ana=cleaned_data.drop('Year',axis=1).groupby('Country Code', as_index=False).mean()
year_ana=cleaned_data.drop('Country Code',axis=1).groupby('Year').mean()
country_ana=country_ana.set_index('Country Code')
#year_ana=year_ana.set_index('Year')
```

# Inferences :

```python
corr=pd.DataFrame()
corr['v_emp']=cleaned_data['Vulnerable employment, female (% of female employment) (modeled ILO estimate)']+cleaned_data['Vulnerable employment, male (% of male employment) (modeled ILO estimate)']
corr['lit']=cleaned_data['Literacy rate, youth (ages 15-24), gender parity index (GPI)']
corr['ch_lab']=(cleaned_data['Children in employment, female (% of female children ages 7-14)']+cleaned_data['Children in employment, male (% of male children ages 7-14)'])
corr['unemp']=cleaned_data['Unemployment, female (% of female labor force) (modeled ILO estimate)']+cleaned_data['Unemployment, male (% of male labor force) (modeled ILO estimate)']
corr.set_index(cleaned_data['Year'])
pd.plotting.scatter_matrix(corr, figsize=(12,10))
plt.show()
corr.corr()
```

### Features of interest :

    1. v_emp : Vulnerable employment (female,male)
    2. li    : Literacy Rate
    3. ch_lab: Child Employment (female,male)
    4. unemp : Unemployment (female,male)

### From the scatter matrix above, and the table of correlation  is evident of chances of child labor affecting the population of vulnerable employees.For which literacy rate reveals a negative correlation.Positive correlation between Literacy Rate and Unemployment highlights the impact of education and knowledge in the domains of emploment and evolution.
